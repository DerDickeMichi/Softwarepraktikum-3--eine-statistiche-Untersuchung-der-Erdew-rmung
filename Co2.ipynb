```python
import numpy as np
import wbgapi as wb
import matplotlib.pyplot as plt

Emissionen_D = wb.data.DataFrame('EN.GHG.CO2.MT.CE.AR5','DEU',range(1970,2023)).to_numpy()[0]
Emissionen_W = wb.data.DataFrame('EN.GHG.CO2.MT.CE.AR5','WLD',range(1970,2023)).to_numpy()[0]
Emissionen_HIC = wb.data.DataFrame('EN.GHG.CO2.MT.CE.AR5','HIC',range(1970,2023)).to_numpy()[0]
Emissionen_EUU = wb.data.DataFrame('EN.GHG.CO2.MT.CE.AR5','EUU',range(1970,2023)).to_numpy()[0]
Bevoelkerung_D = wb.data.DataFrame('SP.POP.TOTL','DEU',range(1970,2023)).to_numpy()[0]
Bevoelkerung_W = wb.data.DataFrame('SP.POP.TOTL','WLD',range(1970,2023)).to_numpy()[0]
Bevoelkerung_HIC = wb.data.DataFrame('SP.POP.TOTL','HIC',range(1970,2023)).to_numpy()[0]
Bevoelkerung_EUU = wb.data.DataFrame('SP.POP.TOTL','EUU',range(1970,2023)).to_numpy()[0]

x=np.linspace(1970,2023,53)

fig,ax1 = plt.subplots()
ax1.plot(x,Emissionen_D, 'b',label='Emission')
ax1.set_ylabel('CO2-Emissionen Deutschlandweit')
ax2 = ax1.twinx()
ax2.plot(x,Bevoelkerung_D, 'r',label='Bevlkerung')
ax2.set_ylabel('Bevlkerungszahl Deutschlandweit')
plt.xlabel('Jahr')
fig.legend()

fig2,ax3 = plt.subplots()
ax3.plot(x,Emissionen_W, 'b',label='Emission')
ax3.set_ylabel('CO2-Emissionen Weltweit')
ax4 = ax3.twinx()
ax4.plot(x,Bevoelkerung_W, 'r',label='Bevlkerung')
ax4.set_ylabel('Bevlkerungszahl Weltweit')
plt.xlabel('Jahr')
fig2.legend()
plt.show()
```

```python
Durchschnitt_D = []
Durchschnitt_W = []
for i in range(0,53):
    Durchschnitt_D.append(Emissionen_D[i]/Bevoelkerung_D[i])
    Durchschnitt_W.append(Emissionen_W[i]/Bevoelkerung_W[i])
    
plt.plot(x,Durchschnitt_D, 'b',label='Deutschlandweit')
plt.plot(x,Durchschnitt_W, 'r',label='Weltweit')
plt.ylabel("Emissionen pro Kopf")
plt.xlabel("Jahr")

plt.legend(loc='upper center')
plt.show()
```
