```python
from meteostat import Point, Daily
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Set the location for Hamburg
hamburg = Point(53.550341, 10.000654)

# Define the time period (1980 - today) as datetime objects
start = datetime(1980, 1, 1)
end = datetime.today()  # Use the current date for the end time

# Retrieve daily weather data for Hamburg
data = Daily(hamburg, start, end)
data = data.fetch()

# Check if data is available
if not data.empty:
    # Add a year column to group by year
    data['year'] = data.index.year

    # Calculate the annual average temperature
    annual_avg_temp = data.groupby('year')['tavg'].mean()

    # Perform linear regression on the annual average temperatures
    years = annual_avg_temp.index
    temps = annual_avg_temp.values
    slope, intercept, r_value, p_value, std_err = stats.linregress(years, temps)

    # Calculate the trendline values
    trendline = slope * years + intercept

    # Plot the data
    plt.figure(figsize=(12, 6))
    plt.plot(annual_avg_temp.index, annual_avg_temp, marker='o', linestyle='-', color='b', label='Jahresdurchschnitt')
    plt.plot(years, trendline, linestyle='--', color='r', label=f'Trendlinie (Steigung = {slope:.2f})')
    plt.title('Jhrliche Durchschnittstemperatur in Hamburg (1980-2024)', fontsize=14)
    plt.xlabel('Jahr', fontsize=12)
    plt.ylabel('Temperatur (\u00b0C)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()

    # Show the plot
    plt.show()

else:
    print("Keine Wetterdaten fr den angegebenen Zeitraum verfgbar.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from meteostat import Point, Daily
from scipy import stats

# Standort Hamburg festlegen
hamburg = Point(53.550341, 10.000654)

# Zeitraum definieren (1980 - heute)
start = datetime(1980, 1, 1)
end = datetime.today()  # Heutiges Datum verwenden

# Wetterdaten fr Hamburg abrufen
data = Daily(hamburg, start, end)
data = data.fetch()

# Prfen, ob Daten verfgbar sind
if not data.empty:
    # Schwellenwert fr Hitzewellen definieren (z. B. tmax > 30C)
    schwellenwert_hitzewelle = 30.0

    # Hitzewellentage identifizieren (tmax > Schwellenwert)
    data['hitzewelle'] = data['tmax'] > schwellenwert_hitzewelle

    # Gruppen fr aufeinanderfolgende Hitzewellentage bilden
    data['hitzewellen_gruppe'] = (data['hitzewelle'] != data['hitzewelle'].shift()).cumsum()

    # Aufeinanderfolgende Hitzewellentage zhlen
    hitzewellen_zaehler = data.groupby('hitzewellen_gruppe').agg({'hitzewelle': 'sum'})

    # Gruppen mit mindestens 3 aufeinanderfolgenden Tagen filtern (Hitzewellen definieren)
    hitzewellen_perioden = hitzewellen_zaehler[hitzewellen_zaehler['hitzewelle'] >= 3]

    # Jahresinformation hinzufgen
    data['jahr'] = data.index.year

    # Anzahl der Hitzewellen pro Jahr zhlen
    hitzewellen_pro_jahr = data.loc[data['hitzewellen_gruppe'].isin(hitzewellen_perioden.index), 'jahr'].value_counts().sort_index()

    # Fehlende Jahre mit 0 auffllen
    hitzewellen_pro_jahr = hitzewellen_pro_jahr.reindex(range(1980, end.year + 1), fill_value=0)

    # Maximale Temperatur whrend der Hitzewellen berechnen
    max_temp_per_group = data.groupby('hitzewellen_gruppe')['tmax'].max()
    max_temp_per_group = max_temp_per_group[hitzewellen_perioden.index]
    hitzewellen_max_temp_pro_jahr = max_temp_per_group.groupby(data.loc[data['hitzewellen_gruppe'].isin(hitzewellen_perioden.index), 'jahr']).max()

    # Fehlende Jahre mit NaN auffllen
    hitzewellen_max_temp_pro_jahr = hitzewellen_max_temp_pro_jahr.reindex(range(1980, end.year + 1), fill_value=np.nan)

    # Lineare Trendanalyse der Hufigkeit von Hitzewellen
    trend_hitzewellen = stats.linregress(hitzewellen_pro_jahr.index, hitzewellen_pro_jahr.values)

    # Plot: Hitzewellen-Hufigkeit mit Trendlinie
    plt.figure(figsize=(14, 6))
    plt.plot(hitzewellen_pro_jahr.index, hitzewellen_pro_jahr.values, label='Hitzewellen pro Jahr', marker='o', color='blue')
    plt.plot(hitzewellen_pro_jahr.index, 
             trend_hitzewellen.intercept + trend_hitzewellen.slope * hitzewellen_pro_jahr.index,
             label=f'Trendlinie (Steigung = {trend_hitzewellen.slope:.2f})', linestyle='--', color='red')
    plt.title('Anzahl der Hitzewellen in Hamburg (1980-2024)', fontsize=14)
    plt.xlabel('Jahr', fontsize=12)
    plt.ylabel('Anzahl der Hitzewellen', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Lineare Trendanalyse der maximalen Temperaturen
    valid_years = ~np.isnan(hitzewellen_max_temp_pro_jahr.values)

    if valid_years.sum() > 1:  # Mindestens 2 Punkte fr eine Regression notwendig
        trend_intensitaet = stats.linregress(
            hitzewellen_max_temp_pro_jahr.index[valid_years],
            hitzewellen_max_temp_pro_jahr.values[valid_years]
        )

        # Plot: Maximale Temperaturen whrend der Hitzewellen mit Trendlinie
        plt.figure(figsize=(14, 6))
        plt.plot(hitzewellen_max_temp_pro_jahr.index, hitzewellen_max_temp_pro_jahr.values, label='Maximale Hitzewellen-Temperatur pro Jahr', marker='o', color='orange')
        plt.plot(hitzewellen_max_temp_pro_jahr.index, 
                 trend_intensitaet.intercept + trend_intensitaet.slope * hitzewellen_max_temp_pro_jahr.index,
                 label=f'Trendlinie (Steigung = {trend_intensitaet.slope:.2f})', linestyle='--', color='red')
        plt.title('Maximale Temperaturen whrend Hitzewellen in Hamburg (1980-2024)', fontsize=14)
        plt.xlabel('Jahr', fontsize=12)
        plt.ylabel('Maximale Temperatur (\u00b0C)', fontsize=12)
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
    else:
        print("Nicht gengend gltige Daten fr die Berechnung der Trendlinie der maximalen Temperaturen.")
else:
    print("Keine Wetterdaten fr den angegebenen Zeitraum verfgbar.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from meteostat import Point, Daily
from scipy import stats

# Standort Hamburg festlegen
hamburg = Point(53.550341, 10.000654)

# Zeitraum definieren (1980 - heute)
start = datetime(1980, 1, 1)
end = datetime.today()  # Heutiges Datum verwenden

# Wetterdaten fr Hamburg abrufen
data = Daily(hamburg, start, end)
data = data.fetch()

# Prfen, ob Daten verfgbar sind
if not data.empty:
    # Schwellenwert fr Hitzewellen definieren (tmax > 30C)
    schwellenwert_hitzewelle = 30.0

    # Hitzewellentage identifizieren (tmax > Schwellenwert)
    data['hitzewelle'] = data['tmax'] > schwellenwert_hitzewelle

    # Gruppen fr aufeinanderfolgende Hitzewellentage bilden
    data['hitzewellen_gruppe'] = (data['hitzewelle'] != data['hitzewelle'].shift()).cumsum()

    # Aufeinanderfolgende Hitzewellentage zhlen
    hitzewellen_zaehler = data.groupby('hitzewellen_gruppe').agg({'hitzewelle': 'sum'})

    # Gruppen mit mindestens 3 aufeinanderfolgenden Tagen filtern (Hitzewellen definieren)
    hitzewellen_perioden = hitzewellen_zaehler[hitzewellen_zaehler['hitzewelle'] >= 3]

    # Jahresinformation hinzufgen
    data['jahr'] = data.index.year

    # Anzahl der Hitzewellen pro Jahr zhlen
    hitzewellen_pro_jahr = data.loc[data['hitzewellen_gruppe'].isin(hitzewellen_perioden.index), 'jahr'].value_counts().sort_index()

    # Fehlende Jahre mit 0 auffllen
    hitzewellen_pro_jahr = hitzewellen_pro_jahr.reindex(range(1980, end.year + 1), fill_value=0)

    # Maximale Temperatur whrend der Hitzewellen berechnen
    max_temp_per_group = data.groupby('hitzewellen_gruppe')['tmax'].max()
    max_temp_per_group = max_temp_per_group[hitzewellen_perioden.index]
    hitzewellen_max_temp_pro_jahr = max_temp_per_group.groupby(data.loc[data['hitzewellen_gruppe'].isin(hitzewellen_perioden.index), 'jahr']).max()

    # Fehlende Jahre mit NaN auffllen
    hitzewellen_max_temp_pro_jahr = hitzewellen_max_temp_pro_jahr.reindex(range(1980, end.year + 1), fill_value=np.nan)

    # Lineare Trendanalyse der Hufigkeit von Hitzewellen
    trend_hitzewellen = stats.linregress(hitzewellen_pro_jahr.index, hitzewellen_pro_jahr.values)

    # Berechnung der jhrlichen Durchschnittstemperaturen
    annual_avg_temp = data.groupby('jahr')['tavg'].mean()

    # Lineare Regression der jhrlichen Durchschnittstemperaturen
    years = annual_avg_temp.index
    temps = annual_avg_temp.values
    slope_temp, intercept_temp, r_value, p_value, std_err = stats.linregress(years, temps)
    trendline_temp = slope_temp * years + intercept_temp

    # Plot: Hitzewellen-Hufigkeit und Temperaturtrends
    fig, ax1 = plt.subplots(figsize=(14, 6))

    # Plot der Hitzewellen-Hufigkeit
    ax1.plot(hitzewellen_pro_jahr.index, hitzewellen_pro_jahr.values, label='Hitzewellen pro Jahr', marker='o', color='blue')
    ax1.plot(hitzewellen_pro_jahr.index, 
             trend_hitzewellen.intercept + trend_hitzewellen.slope * hitzewellen_pro_jahr.index,
             label=f'Trendlinie Hitzewellen (Steigung = {trend_hitzewellen.slope:.2f})', linestyle='--', color='red')
    ax1.set_xlabel('Jahr', fontsize=12)
    ax1.set_ylabel('Anzahl der Hitzewellen', fontsize=12)
    ax1.tick_params(axis='y', labelcolor='blue')

    # Sekundre y-Achse fr Temperaturtrend
    ax2 = ax1.twinx()
    ax2.plot(years, annual_avg_temp, marker='o', linestyle='-', color='green', label='Jhrliche Durchschnittstemperatur')
    ax2.plot(years, trendline_temp, linestyle='--', color='orange', label=f'Trendlinie Temperatur (Steigung = {slope_temp:.2f})')
    ax2.set_ylabel('Temperatur (C)', fontsize=12)
    ax2.tick_params(axis='y', labelcolor='green')

    # Titel und Legende
    plt.title(f'Hitzewellen und Jhrliche Durchschnittstemperatur in Hamburg (1980-{end.year})', fontsize=14)
    fig.tight_layout()
    ax1.grid(True, linestyle='--', alpha=0.7)
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    # Plot anzeigen
    plt.show()

else:
    print("Keine Wetterdaten fr den angegebenen Zeitraum verfgbar.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from meteostat import Point, Daily

# Standort Hamburg festlegen
hamburg = Point(53.550341, 10.000654)

# Zeitraum definieren (1980 - 2023)
start = datetime(1980, 1, 1)
end = datetime.today()

# Wetterdaten fr Hamburg abrufen
data = Daily(hamburg, start, end)
data = data.fetch()

# Prfen, ob Daten verfgbar sind
if not data.empty:
    # Jahresinformation hinzufgen
    data['jahr'] = data.index.year

    # Ergebnisse fr verschiedene Schwellenwerte berechnen
    schwellenwerte = [30]  # Verschiedene Temperaturgrenzen
    ergebnisse = []

    for schwellenwert in schwellenwerte:
        # Hitzewellentage und markierte Hitzewellen definieren
        data['ueber_schwellenwert'] = data['tmax'] > schwellenwert
        data['hitzewellen_gruppe'] = (data['ueber_schwellenwert'] != data['ueber_schwellenwert'].shift()).cumsum()
        hitzewellen_zaehler = data.groupby('hitzewellen_gruppe').agg({'ueber_schwellenwert': 'sum'})
        hitzewellen_perioden = hitzewellen_zaehler[hitzewellen_zaehler['ueber_schwellenwert'] >= 3]
        data['hitzewelle_markiert'] = data['hitzewellen_gruppe'].isin(hitzewellen_perioden.index)

        # Jahresweise Ergebnisse sammeln
        jahres_ergebnisse = data.groupby('jahr').apply(
            lambda x: {
                'tage_ueber_schwellenwert': x['ueber_schwellenwert'].sum(),
                'hitzewellen': x['hitzewelle_markiert'].any(),
                'wahrscheinlichkeit': (
                    x['hitzewelle_markiert'].any() / x['ueber_schwellenwert'].sum()
                    if x['ueber_schwellenwert'].sum() > 0 else 0
                ) * 100  # Wahrscheinlichkeit in Prozent
            }
        )
        jahres_ergebnisse = pd.DataFrame(list(jahres_ergebnisse), index=jahres_ergebnisse.index)
        jahres_ergebnisse['schwellenwert'] = schwellenwert
        ergebnisse.append(jahres_ergebnisse)

    # Alle Ergebnisse in einem DataFrame zusammenfassen
    ergebnisse = pd.concat(ergebnisse).reset_index()
    ergebnisse.rename(columns={'index': 'jahr'}, inplace=True)

    # Visualisierung
    plt.figure(figsize=(16, 8))
    for schwellenwert in schwellenwerte:
        schwellenwert_data = ergebnisse[ergebnisse['schwellenwert'] == schwellenwert]
        plt.bar(
            schwellenwert_data['jahr'],
            schwellenwert_data['wahrscheinlichkeit'],
            label=f'Schwellenwert {schwellenwert}C',
            alpha=0.7
        )
    plt.title('Wahrscheinlichkeit einer Hitzewellen bei Schwellenwerten 30C (1980-2024)', fontsize=14)
    plt.xlabel('Jahr', fontsize=12)
    plt.ylabel('Wahrscheinlichkeit einer Hitzewelle (%)', fontsize=12)
    plt.legend(title="Schwellenwerte")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("Keine Wetterdaten fr den angegebenen Zeitraum verfgbar.")
```
